### Motivation 
- Tuberculosis (TB) is a lung disease causing ~1.5 million deaths/year and is caused by the bacterium *Mycobacterium tuberculosis*
- 20% of cases exhibit resistance to one or more drugs
- Understanding the mechanism of action of new anti-TB drugs helps speed up drug discovery

- In collaboration with the Broad Institute of MIT and Harvard, we developed an approach to predict the mechanism of action of a new drug

### Problem

- Predict mechanism of action of a drug by measuring how effective a drug is on different kinds (strains) of tuberculosis.
- Secondary goal: Reduce cost of the experiment by minimising the number of strains required to make predictions

The input data for this project is
- A set of ~425 *M. tuberculosis* strains exposed to
- 80 different drugs belonging to 
- 6 *known* drug classes (A-F)
- Each drug is used at multiple different concentrations
- The processed data is of the form Drug-Concentration as rows, and Strains as columns


### Objectives/Goals

1. Clean sequencing reads and convert them to a matrix of Drug-Concentration x Strains (https://git.io/JfsKh)
2. Analyze the quality of the data. Specifically, is there missing data, and if so why? (https://git.io/Jfs6v)
3. Fit a machine learning model to predict drug classes from drug-strain profiles (https://git.io/Jfs6T)
4. Find a minimal set of features (a.k.a strains) that are sufficient to predict drug classes (https://git.io/Jfs6T)
