### Motivation for the project
- Tuberculosis is a lung disease that causes the greatest number of deaths/year due to an infectious agent. 
- It is caused by the bacterium *Mycobacterium tuberculosis*
- In 2018, the disease infected ~10 million people and caused ~1.5 million deaths
- 20% of cases exhibit resistance to one or more drugs
- Understanding the mechanism of action of new anti-TB drugs helps speed up drug discovery

- In collaboration with the Broad Insititute of MIT and Harvard, we developed a chemical-genetic approach to predict the mechanism of action of a new drug

The input data for this project is
- A set of ~500 *M. tuberculosis* strains exposed to
- 80 different drugs belonging to 
- 6 *known* drug classes (A-F)
- Each drug is used at multiple different concentrations
- The processed data is of the form Drug-Concentration as rows, and Strains as columns


### Objectives/Goals

1. Clean sequencing reads and convert them to a matrix of Drug-Concentration x Strains
2. Analyze the quality of the data. Specifically, is there missing data, and if so why? **#Goal1**
3. Goal: Fit a machine learning model to predict drug classes from drug-strain profiles **#Goal2**
4. Goal: Find a minimal set of features (a.k.a strains) that are sufficient to predict drug classes **#Goal3**
