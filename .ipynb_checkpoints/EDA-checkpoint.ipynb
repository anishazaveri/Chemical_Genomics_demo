This notebook looks at possible causes for bad PCRs in the experiment. First we establish that the number of bad PCRs is higher in this experiment compared to a previous iteration and we also breakdown the number of bad PCRs by drug. 

Next, we test if the PCRs are poor because of lack of input (look at OD plots) or because of jackpotting. We find that while jackpotting exists, it cannot explain the poor PCR quality. We conclude that likely the errors are due to technical PCR difficulties and decide to standardize the PCR conditions.

## Jackpotting?

Is there jackpotting? In other words are some strains overly represented even before drug treatment. This might explain why the PCRs are bad. 

How does the jackpotting in this HypoIII experiment compare with different experiments? Specifically how does this compare to the 160 strain non hypo mixes? (exp8)

```python
%reset -f
```

```python
import pandas as pd
import numpy as np
import os
import pathlib
import itertools
import seaborn as sns
from plotnine import *
%matplotlib inline
```

```python
hypo3counts = pd.read_csv(pathlib.Path.cwd().joinpath('Analysis','3-Merge_counts', 'Counts_merged.csv'))
hypoIcounts = pd.read_csv(pathlib.Path.cwd().parents[0].joinpath('HypoI', 'Analysis','Path-A', '3-Merge_counts', 'Counts_merged.csv'))
exp8_counts = pd.read_csv(pathlib.Path.cwd().joinpath('Raw_data', 'Exp8_counts.csv'))
```

Get strains of expt 8, hypoI and hypoIII, order them in descending order of counts and calculate cumulative sum

```python
exp8 = exp8_counts[exp8_counts['xMIC']=='0.000xMIC']
exp8 = exp8.filter(regex='/', axis=1)
exp8 = exp8.melt(value_name='RC', var_name='Strain')
exp8 = exp8.groupby('Strain', as_index=False)['RC'].mean().sort_values(by='RC', ascending=False)
exp8['cumsum'] = exp8['RC'].cumsum().sort_values(ascending=False)
exp8['cumsum'] = exp8['cumsum']/sum(exp8['RC'])
exp8 = exp8.reset_index(drop=True)
```

```python
def generate_counts_df_hypo(data):
    data = data[data['Conc']=='0.000xMIC']
    data = data.groupby('Strain', as_index=False)['RC'].mean().sort_values(by='RC', ascending=False)
    data['cumsum'] = data['RC'].cumsum().sort_values(ascending=False)
    data['cumsum'] = data['cumsum']/sum(data['RC'])
    data = data.reset_index(drop=True)
    #print(data.head(10))
    return data 
```

```python
hypoI = generate_counts_df_hypo(hypoIcounts)
hypo3 = generate_counts_df_hypo(hypo3counts)
```

```python
cumsums = pd.DataFrame({'exp8': exp8['cumsum'], 'hypoI':hypoI['cumsum'], 'hypoIII':hypo3['cumsum']})
cumsums.tail()
```

```python
# melt to make compatible with ggplot
cumsums_m = cumsums.melt(var_name='Expt', value_name='cumsum')
cumsums_m.head()
```

```python
cumsums_m['Expt'].value_counts()
```

```python
# generate x-coordinate
cumsums_m['x_coord_plot'] = list(range(0,438))*3
```

```python
(ggplot(data=cumsums_m, mapping=aes(x='x_coord_plot', y='cumsum'))
 +geom_point(size=0.5)
 +facet_wrap('Expt')
 +geom_abline(intercept=0, slope=1/438, linetype='dashed', color='red')
 +labs(y='Fraction cumulative reads', x='Number of strains')
 +theme_light()
 +theme(figure_size=(9,3), strip_text =element_text(color="black"))
 +ylim(0,1))
```

^^
Ideally each strain should have an equal contribution to the mix, and the cumulative reads will lie on the dotted red line. Importantly, this is not true. In HypoI and HypoIII about 6 strains contribute to 60% of the reads. In experiment 8, there is some jackpotting as well but this is much lower than in the hypo mixes

This may explain why some PCRs are bad. Specifically in the bad PCRs maybe the extent of jackpotting is higher. Will check in the next section

## Compare jackpot strains between good and bad PCRs

Which strains are jackpotting?

```python
hypo3.head()
```

```python
pcrs = pd.DataFrame(pd.read_csv(pathlib.Path.cwd().joinpath('Analysis','5-QC', 'Sel_ODs_drugs_counts_QC_PCR_flagged.csv')))
```

```python
strain_cols = pcrs.filter(regex='\.', axis=1).columns
pcrs[strain_cols] = np.log10(pcrs[strain_cols])
```

```python
(ggplot(data=pcrs, mapping=aes(x='below_threshold', y='strain.240', color='quality'))
 +geom_point()
 +theme_light()
 +theme(figure_size=(3,3))
 +scale_color_manual(values={'bad':'#d95f02', 'good':'#1b9e77'}, name='PCR quality')
 +labs(y='log10(RC) strain.240', x='Number of strains below threshold'))
```

```python
(ggplot(data=pcrs, mapping=aes(x='below_threshold', y='strain.424', color='quality'))
 +geom_point()
 +theme_light()
 +theme(figure_size=(3,3))
 +scale_color_manual(values={'bad':'#d95f02', 'good':'#1b9e77'}, name='PCR quality')
 +labs(y='log10(RC) strain.424', x='Number of strains below threshold'))
```

For the top two strains, we see that the PCRs which are bad have low counts. In other words, the jackpotting strains do not over-represent counts in the bad PCRs.

Therefore, while jackpotting may be an issue for obtaining enough sequencing coverage of the other strains, it cannot explain the failed PCRs

It is likely that the failed PCRs arise from technical issues during the PCR itself.
