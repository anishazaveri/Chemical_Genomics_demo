```python
import pandas as pd
import pathlib
```

```python
df = pd.read_csv(pathlib.Path.cwd().joinpath('Analysis','8-Clean_for_ML', 'HypoIII_for_ML_encoded.csv'))
```

```python
df.head()
```

```python
#Get concentration ranks. 
#This allows us to choose the top three concentrations that have mean OD above the desired cutoff of 0.3
df['num_conc'] = df['Conc'].str.replace("xMIC", "").astype('float')
df['Conc_rank'] = df.groupby('Drug')['num_conc'].rank(method='dense')
```
